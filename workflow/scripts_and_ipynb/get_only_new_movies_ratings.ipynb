```python
path = 'small/ml-latest-small/'
```

```python
import pandas as pd
import numpy as np
```

```python
actual_data = pd.read_csv(path+'final_data_to_scrap.csv')
```

```python
movie_id_list = list(actual_data.T.loc['movieId'])
```

```python
ratings = pd.read_csv(path+'ratings.csv')
```

```python
ratings
```

```python
ratings.index
```

```python
to_drop = []
for i in list(ratings.index):
    if int(ratings.T.loc['movieId'][i]) not in movie_id_list:
        to_drop.append(i)
        
    
```

```python
len(to_drop)
```

```python
len(ratings) - len(to_drop)
```

```python
final_ratings = ratings.drop(to_drop)
```

```python
final_ratings
```

```python
final_ratings.to_csv("final_ratings.csv")
```
